```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri
strike = 245           # Strike dell'opzione call
premio = 1.53            # Premio per azione dell'opzione
moltiplicatore = 100     # Numero di azioni controllate dalla call
quantit_short = 50      # Numero di azioni vendute short
prezzo_vendita = 245

# Range dei prezzi dell'azione a scadenza
S = np.linspace(230, 260, 100)

# Payoff della call: max(S - strike, 0)*numero azioni - premio*numero azioni
payoff_call = np.maximum(S - strike, 0) * moltiplicatore - premio * moltiplicatore

# Payoff della posizione short: (prezzo_vendita - S)*quantit
payoff_short = (prezzo_vendita - S) * quantit_short

# Payoff aggregato
payoff_totale = payoff_call + payoff_short

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(S, payoff_call, label='Call', linestyle='--')
plt.plot(S, payoff_short, label='Short 50 Azioni', linestyle=':')
plt.plot(S, payoff_totale, label='Payoff Aggregato', linewidth=3)  # evidenziato con linea pi spessa
plt.xlabel('Prezzo Sottostante a Scadenza')
plt.ylabel('Profitto / Perdita')
plt.title('Delta Hedging Payoff a scadenza')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
