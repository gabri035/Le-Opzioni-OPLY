```python
import numpy as np
import matplotlib.pyplot as plt

# Strike prices
K1 = 60  # Strike della Call comprata
K2 = 50  # Strike della Call venduta

# Premi delle opzioni
premio_call_acquistata = 2
premio_call_venduta = 5

# Prezzi del sottostante alla scadenza
prezzi_sottostante = np.linspace(40, 70, 200)

# Payoff della Call acquistata (K1)
payoff_call_acquistata = np.maximum(prezzi_sottostante - K1, 0) - premio_call_acquistata

# Payoff della Call venduta (K2)
payoff_call_venduta = premio_call_venduta - np.maximum(prezzi_sottostante - K2, 0)

# Payoff totale della strategia
payoff_totale_spread = payoff_call_acquistata + payoff_call_venduta

# Grafico
plt.figure(figsize=(10, 6))
plt.plot(prezzi_sottostante, payoff_call_acquistata, '-', label='Call Comprata (K1)')
plt.plot(prezzi_sottostante, payoff_call_venduta, '-', label='Call Venduta (K2)')
plt.plot(prezzi_sottostante, payoff_totale_spread, label='Spread Verticale Call')
plt.axhline(0, color='black', linewidth=0.7)

# Annotazioni strike
plt.text(K1, 0, 'K2', horizontalalignment='right', verticalalignment='bottom', fontsize=12, color='blue')
plt.text(K2, 0, 'K1', horizontalalignment='right', verticalalignment='bottom', fontsize=12, color='red')

plt.title('Payoff Bear Call Spread')
plt.xlabel('Prezzo Sottostante')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Strike prices
K1 = 50  # Strike della Put venduta
K2 = 60  # Strike della Put comprata

# Premi delle opzioni
premio_put_venduta = 2
premio_put_acquistata = 5

# Prezzi del sottostante alla scadenza
prezzi_sottostante = np.linspace(40, 70, 200)

# Payoff della Put venduta (K1)
payoff_put_venduta = premio_put_venduta - np.maximum(K1 - prezzi_sottostante, 0)

# Payoff della Put acquistata (K2)
payoff_put_acquistata = np.maximum(K2 - prezzi_sottostante, 0) - premio_put_acquistata

# Payoff totale della strategia (Bull Put Spread)
payoff_totale_bull_put = payoff_put_venduta + payoff_put_acquistata

# Grafico
plt.figure(figsize=(10, 6))
plt.plot(prezzi_sottostante, payoff_put_venduta, '-', label='Put Venduta (K1)')
plt.plot(prezzi_sottostante, payoff_put_acquistata, '-', label='Put Comprata (K2)')
plt.plot(prezzi_sottostante, payoff_totale_bull_put, label='Bull Put Spread')
plt.axhline(0, color='black', linewidth=0.7)
# Annotazioni degli strike K1 e K2
plt.text(K1, 0, 'K1', ha='center', va='bottom', fontsize=12, color='blue')
plt.text(K2, 0, 'K2', ha='center', va='bottom', fontsize=12, color='red')

plt.title('Payoff Bear Put Spread')
plt.xlabel('Prezzo Sottostante')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)
plt.show()
```
