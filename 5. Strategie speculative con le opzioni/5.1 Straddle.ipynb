```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri
K = 50              # Strike
premio = 2.5        # Premio ricevuto per ciascuna opzione
S = np.linspace(30, 70, 400)  # Range dei possibili prezzi del sottostante

# Calcolo dei payoff per la vendita (short) delle opzioni
payoff_call_venduta = premio - np.maximum(S - K, 0)  # Vendita di una call
payoff_put_venduta = premio - np.maximum(K - S, 0)     # Vendita di una put
payoff_totale = payoff_call_venduta + payoff_put_venduta

# Creazione del grafico
plt.figure(figsize=(8,4))
plt.plot(S, payoff_call_venduta, label="Payoff Call Venduta")
plt.plot(S, payoff_put_venduta, label="Payoff Put Venduta")
plt.plot(S, payoff_totale, label="Payoff Aggregato", linewidth=2, color="green")
plt.xlabel("Prezzo del Sottostante")
plt.ylabel("Payoff")
plt.title("Short Straddle")
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(K, color='black', linewidth=0.5, linestyle="--", label="Strike = 50")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri
K = 50              # Strike
premio = 2.5        # Premio ricevuto per ciascuna opzione
S = np.linspace(30, 70, 400)  # Range dei possibili prezzi del sottostante

# Calcolo dei payoff per la vendita (short) delle opzioni
payoff_call_venduta =  np.maximum(S - K, 0) - premio  # Vendita di una call
payoff_put_venduta = np.maximum(K - S, 0)  - premio   # Vendita di una put
payoff_totale = payoff_call_venduta + payoff_put_venduta

# Creazione del grafico
plt.figure(figsize=(8,4))
plt.plot(S, payoff_call_venduta, label="Payoff Call Venduta")
plt.plot(S, payoff_put_venduta, label="Payoff Put Venduta")
plt.plot(S, payoff_totale, label="Payoff Aggregato", linewidth=2, color="green")
plt.xlabel("Prezzo del Sottostante")
plt.ylabel("Payoff")
plt.title("Long straddle")
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(K, color='black', linewidth=0.5, linestyle="--", label="Strike = 50")
plt.legend()
plt.grid(True)
plt.show()
```
