```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri della strategia
prezzo_corrente = 100
strike_call = 110      # Strike della call OTM
strike_put = 90        # Strike della put OTM
premium_call = 3     # Premio pagato per la call
premium_put = 2.5      # Premio pagato per la put

# Intervallo dei possibili prezzi del sottostante
S = np.linspace(75, 125, 150)

# Calcolo del payoff netto per ciascuna opzione
# Si sottrae il premio per ottenere il payoff netto
net_payoff_call = np.maximum(S - strike_call, 0) - premium_call
net_payoff_put = np.maximum(strike_put - S, 0) - premium_put

# Payoff totale della strategia Strangle Long
payoff_strangle = net_payoff_call + net_payoff_put

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(S, net_payoff_call, label='Payoff Call', linestyle='-', linewidth=2)
plt.plot(S, net_payoff_put, label='Payoff Put', linestyle='-', linewidth=2)
plt.plot(S, payoff_strangle, label='Payoff Strangle Long', color='green', linewidth=3)

# Linea orizzontale a 0 per evidenziare il break-even
plt.axhline(0, color='black', linewidth=2, linestyle='-')
plt.axvline(prezzo_corrente, color='blue', linestyle=':', linewidth=1, label='Prezzo Corrente')

plt.xlabel('Prezzo del Sottostante')
plt.ylabel('Payoff')
plt.title('Strangle Long')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri della strategia Short Strangle
prezzo_corrente = 100
strike_call = 110      # Strike della call venduta (OTM)
strike_put = 90        # Strike della put venduta (OTM)
premium_call = 3     # Premio incassato per la call
premium_put = 2.5      # Premio incassato per la put

# Definizione dell'intervallo dei possibili prezzi del sottostante
S = np.linspace(75, 125, 125)

# Calcolo del payoff netto per ciascuna opzione venduta:
# Per la short call: il payoff è dato dal premio incassato meno l'eventuale esborso se l'opzione finisse in the money.
payoff_short_call = premium_call - np.maximum(S - strike_call, 0)
# Per la short put: il payoff è dato dal premio incassato meno l'eventuale esborso se l'opzione finisse in the money.
payoff_short_put = premium_put - np.maximum(strike_put - S, 0)

# Calcolo del payoff totale della strategia Short Strangle
payoff_total = payoff_short_call + payoff_short_put

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(S, payoff_short_call, label='Payoff Short Call', linewidth=2)
plt.plot(S, payoff_short_put, label='Payoff Short Put', linewidth=2)
plt.plot(S, payoff_total, label='Payoff Totale Short Strangle', color='green', linewidth=3)


# Linee di riferimento per facilitare l'interpretazione
plt.axhline(0, color='black', linewidth=2)  # asse dei profitti/perdite a 0
plt.axvline(prezzo_corrente, color='blue', linestyle='-', linewidth=1, label='Prezzo Corrente')

# Etichette e titolo
plt.xlabel('Prezzo del Sottostante')
plt.ylabel('Payoff')
plt.title('Strangle Short')
plt.legend()
plt.grid(True)
plt.show()
```
